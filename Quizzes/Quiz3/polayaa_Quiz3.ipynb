````
from statsmodels.graphics.mosaicplot import mosaic

filtered_data = titanic[['Sex', 'Survived']]

# Calculate proportions
grouped_data = filtered_data.groupby(['Sex', 'Survived']).size().unstack()
grouped_data = grouped_data.div(grouped_data.sum(axis=1), axis=0)

# Change the order of columns
grouped_data = grouped_data[[1, 0]]

# Rename columns
grouped_data.columns = ['Survived', 'Died']

# Create a mosaic plot
mosaic(grouped_data.stack(), axes_label=False, title='Survival rate by boarding port')

# Show the plot
plt.show()

````

The chart clearly illustrates the effect of the gender-based priority for lifeboat access ("women and children first"). Survival was highly dependent on being female.

````
from statsmodels.graphics.mosaicplot import mosaic

filtered_data = titanic[['Pclass', 'Survived']]

# Calculate proportions
grouped_data = filtered_data.groupby(['Pclass', 'Survived']).size().unstack()
grouped_data = grouped_data.div(grouped_data.sum(axis=1), axis=0)

# Change the order of columns
grouped_data = grouped_data[[1, 0]]

# Rename columns
grouped_data.columns = ['Survived', 'Died']

# Create a mosaic plot
mosaic(grouped_data.stack(), axes_label=False, title='Survival rate by passenger class')

# Show the plot
plt.show()
````

First-Class passengers had the highest survival rate (around 61.9% overall), followed by Second-Class (around 42.9%), and Third-Class (around 25.5%), primarily due to location on the ship and the "women and children first" protocol. Since the group from Cherbourg had a higher concentration of the wealthy, high-fare passengers who were closer to the lifeboats, their overall survival rate was elevated.

````
filtered_data = titanic[['Embarked', 'Survived']]

# Calculate proportions
grouped_data = filtered_data.groupby(['Embarked', 'Survived']).size().unstack()
grouped_data = grouped_data.div(grouped_data.sum(axis=1), axis=0)

# Change the order of columns
grouped_data = grouped_data[[1, 0]]

# Rename columns
grouped_data.columns = ['Survived', 'Died']

# Create a mosaic plot
mosaic(grouped_data.stack(), axes_label=False)

# Show the plot
plt.show()
````

The port of embarkation appears to correlate with survival, likely due to the economic class distribution of passengers at each port (as established in the previous analysis). The treemap shows the following proportions of Died vs. Survived for each port:

* C (Cherbourg): This port shows the most balanced proportion of Died to Survived; the "C Survived" area is large relative to the "C Died" area. This suggests the highest survival rate of the three ports.

* Q (Queenstown): This port shows a higher proportion of Died than Survived, though the "Q Survived" area is still substantial.

* S (Southampton): This port shows the largest overall proportion of Died compared to Survived among the three ports. The "S Died" area is the largest single category, indicating the lowest overall survival rate, likely due to the large number of Third-Class passengers boarding there .





````
import matplotlib.pyplot as plt

# Group by Embarked and Pclass and count the occurrences
embarked_pclass_counts = titanic.groupby(['Embarked', 'Pclass']).size().unstack()

# Plotting a stacked bar chart
embarked_pclass_counts.plot(kind='bar', stacked=True, figsize=(10, 6))

# Adding labels and title
plt.title('Distribution of Passenger Class by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)
plt.legend(title='Passenger Class')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
````

````
import matplotlib.pyplot as plt
import numpy as np

# Group by Embarked and Pclass and count the occurrences
embarked_pclass_counts = titanic.groupby(['Embarked', 'Pclass']).size().unstack()

# Calculate the number of survivors for each embarkation port
survivors_by_embarked = titanic.groupby('Embarked')['Survived'].sum()

# Plotting a stacked bar chart
ax = embarked_pclass_counts.plot(kind='bar', stacked=True, figsize=(10, 6))

# Ensure the order of ports in survivors_by_embarked matches the order in embarked_pclass_counts
survivors_by_embarked_ordered = survivors_by_embarked.reindex(embarked_pclass_counts.index)

# Plotting the number of survivors as a line on the primary y-axis
ax.plot(np.arange(len(embarked_pclass_counts.index)), survivors_by_embarked_ordered.values, marker='o', color='red', label='Number of Survivors')

# Adding labels and title
plt.title('Distribution of Passenger Class and Survivors by Embarkation Port')
ax.set_xlabel('Embarkation Port')
ax.set_ylabel('Number of Passengers / Survivors') # Update y-axis label
ax.tick_params(axis='x', rotation=0)
ax.legend(title='Passenger Class', loc='upper left')
# Combine legends
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, title='Legend', loc='upper left')

ax.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
````
